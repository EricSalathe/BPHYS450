<h1>Solving ODEs with scipy</h1>

A general ODE can be written in the form:

$\frac{dy}{dt}=f\left(t,y\right);\:y\left(t_o\right)=y_o$

For a particular problem, we can specify the right hand side $f(t,y)$ and the initial condition $y(t_o)=y_o$. 

To solve numerically in python, we define the function f(t,y) and use the scipy odeint function to "integrate" and obtain a solution. For the simple ODE:

$\frac{dy}{dt}=-ky;\:y\left(t_o=0\right)=5,\:\:\:\:\:k=0.1$

We can use the python function 



```python
# function that returns dy/dt
def rate(t,y):
    k = 0.1
    dydt = -k*y
    return dydt
```

```python
# initial condition
t0 = 0
y0 = 5
```

We will use SciPy https://scipy.github.io/devdocs/tutorial/general.html to solve the ODE using the odeint function. There are other methods available, but odeint is fast and simple. 
https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.odeint.html

[The solve_ivp method is more sophisticated and generally preferred but requires coding issues I don't want to get into... Feel free to read up on this and use it if you'd like.]

To use odeint we need to provide the values of the independent variable (t) where we want the solution y(t), starting with t0. Note that this does not affect the numerics, just the values where we will have a solution. If you just want the solution at the final time, this can be a 2-element array:



```python
# time points to solve for
tfinal = 10
npoints = 21
t = linspace(t0,tfinal,npoints)
```

Finally, import the module odeint and run it. It just needs the name of the function giving the derivative, the initial value for the solution, and the array of values for independent variable where we want the solution:

```python
from scipy.integrate import odeint
y = odeint(rate,y0,t, tfirst=True) # note tfirst flag to match form of function

# graph the result
plot(t,y)
```
