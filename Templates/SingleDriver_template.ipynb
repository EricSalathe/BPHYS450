## Single Driver Model

We have the acceleration for a single car from the IDM model as 
<br>
    $\frac{dv}{dt}=a\left[1\:-\:\left(\frac{v}{v_0}\right)^{\delta}\:-\left(\frac{s^{\ast}\left(v,\Delta v\right)}{s}\right)^2\right]$
<br>
where <br> 
$s^{\ast}\left(v,\Delta v\right)=s_0+vT+\frac{v\Delta v}{2\sqrt{ab}}$ <br>
<br>
If we model a single car following a car ahead moving at consant speed $v_{a}$ and starting at position $x_{a0}$, then the car ahead has position $x_a = x_{a0} + v_a t$. We can then compute $\Delta v = v - v_a$ and $s = x_a - L - x$.
