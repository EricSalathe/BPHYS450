# Eigenvalues

The eigenvalues of a matrix are found using the eig function in numpy.linalg. It outputs a 1-d array of eigenvalues and a 2-d array of eigenvectors, one eigenvector in each column. Use the matrix <br>
$A=\begin{bmatrix}
5/4 & 0\\
0 & 3/4
\end{bmatrix}$
<br>

Find and solve the characteristic polynomial $det\left(A\:-\lambda I\right)=0$ to get the eigenvalues

Choose one and solve $\left(A-\lambda I\right)x\:=0$ to find the corresponding eigenvector x.

Visit the web site below and enter this matrix and find the eigenvector eigenvalues graphically
 https://www.geogebra.org/m/JP2XZpzV  

<br>
In python, this is done as follows:

```python
from numpy.linalg import eig
A = array([[5/4,0],[0,3/4]])
eigvalues, eigenvectors = eig(A)

l1 = eigvalues[0]
v1 = eigenvectors[:,0]

l2 = eigvalues[1]
v2 = eigenvectors[:,1]

print('First  pair:',l1, v1)
print('Second pair:',l2, v2)
```

We can check this by computing $A v_1 - \lambda_1 v_1 = 0$ and $A v_2 - \lambda_2 v_2 = 0$.

```python
Av =  matmul(A,v1) # or A@v1
lv = l1*v1
print(Av, lv, Av-lv)
```

```python
Av = matmul(A,v2)
lv = l2*v2
print(Av, lv, Av-lv)
```
